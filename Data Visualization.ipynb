```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Housing.csv")
```

```python
c= []
n=[]
for a in df.columns:
    if df[a].dtype == 'object':
        c.append(a)
    else:
        n.append(a)
print(c)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for e in c:
    df[e] = LE.fit_transform(df[e])
df.head()
```

```python
x= df['price']
y= df['area']
plt.plot(x,y)
```
